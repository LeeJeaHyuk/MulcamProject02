```python
# 비율 저장
import os
import pandas as pd

def died_rate(column_name):
    # 폴더 생성
    try:
        if not os.path.exists('EDAdata'):
            os.makedirs('EDAdata')
    except OSError:
        print ('Error: Creating directory. EDAdata')

    # 파일 경로
    file_path = './EDAdata/' + 'died_rate' + '.csv'

    if os.path.isfile(file_path): # 파일 존재

        entire = column_name + '_entire'
        died = column_name + '_died'
        rate =column_name + '_rate'

        df_rate_new = pd.DataFrame(columns=['index',entire,died,rate])
        
        for index in range(len(df_died.groupby(column_name))):
            try:
                df_d0 = df_died.groupby(column_name).count()['PassengerId'][index] 
                df_0 = df.groupby(column_name).count()['PassengerId'][index]
                df_rate_new = df_rate_new.append(pd.DataFrame([[index,df_0,df_d0,round(df_d0/df_0*100,2)]], columns=['index',entire,died,rate]), ignore_index=True)
            except:
                pass
        
        df_rate = pd.read_csv(file_path)
        df_concat = pd.concat([df_rate,df_rate_new],axis=1)
        df_concat.to_csv(file_path, index=False)

    else: # 파일 존재하지 않음
        entire = column_name + '_entire'
        died = column_name + '_died'
        rate =column_name + '_rate'

        df_rate = pd.DataFrame(columns=['index',entire,died,rate])

        for index in range(len(df_died.groupby(column_name))):
            try:
                df_d0 = df_died.groupby(column_name).count()['PassengerId'][index] 
                df_0 = df.groupby(column_name).count()['PassengerId'][index]
                df_rate = df_rate.append(pd.DataFrame([[index,df_0,df_d0,round(df_d0/df_0*100,2)]], columns=['index',entire,died,rate]), ignore_index=True)
            except:
                pass

        df_rate.to_csv(file_path, index=False)
        

# died_rate('VRDeck')
```
