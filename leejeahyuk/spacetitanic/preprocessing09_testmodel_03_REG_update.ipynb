```python
import numpy as np
import pandas as pd
# update 순서
train = pd.read_csv('./kaggle/input/spaceship-titanic/train.csv')
# df = pd.read_csv('./kaggle/input/spaceship-titanic/test_Destination_update.csv')

# Age 61세보다 낮은 값의 중앙값으로 처리
df = pd.read_csv('./kaggle/input/spaceship-titanic/test_Age_update.csv')
```

```python
df.isnull().sum()
```

```python
df.loc[df['RoomService'].isna() == True, 'RoomService'] = df['RoomService'].median()
df.loc[df['FoodCourt'].isna() == True, 'FoodCourt'] = df['FoodCourt'].median()
df.loc[df['ShoppingMall'].isna() == True, 'ShoppingMall'] = df['ShoppingMall'].median()
df.loc[df['Spa'].isna() == True, 'Spa'] = df['Spa'].median()
df.loc[df['VRDeck'].isna() == True, 'VRDeck'] = df['VRDeck'].median()
```

```python
# df.isnull().sum()
df.to_csv('./kaggle/input/spaceship-titanic/test_update02.csv',index=False)
```
