```python
import numpy as np
import pandas as pd

df = pd.read_csv('./kaggle/input/spaceship-titanic/train_Destination_update.csv')
df_train = df
```

```python
df_train.isnull().sum()
```

```python
# df_FoodCourt =  df_train중에서 FoodCourt null값이 존재하는 열만 추출 
# -> 훈련된 모델의 값으로 다시 채울 예정
df_FoodCourt = df_train[df_train['FoodCourt'].isnull()]
#df_train 중에서 FoodCourt 의 null값이 존재하는 열만 추출 -> 추후 훈련할 데이터프레임
df_train01 = df_train.dropna(subset=['FoodCourt'])
# df_train
df_train[df_train['FoodCourt'].isnull()]
```

```python
# train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_train01.drop(['PassengerId','FoodCourt','Name'],axis=1),df_train01['FoodCourt'], test_size=0.2, random_state=42)

scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
X_train.shape
y_train.shape
X_test.shape
y_test.shape
```

```python
#XGBoost1
import xgboost as xgb
import pandas as pd
import seaborn as sns
from xgboost import XGBRegressor
from sklearn.metrics import accuracy_score

model = xgb.XGBRegressor(early_stopping_rounds=10)
eval_set = [(X_test, y_test)]
model.fit(X_train, y_train, eval_set=eval_set, verbose=False)#, verbose=True

pred = model.predict(X_test)
# acc = accuracy_score(y_test,pred)
# acc

# from sklearn.metrics import confusion_matrix
# import scikitplot as skplt

# pred = model.predict(X_test)
# # 실제 , 예측
# cm = confusion_matrix(y_test, pred)
# print('confusion_matrix : \n', cm)

# skplt.metrics.plot_confusion_matrix(y_test,pred,figsize=(8,6))
# plt.show()

# cl_report = metrics.classification_report(y_test,pred)
# print('리포트:\n',cl_report)
```

```python
# df_01 은 df_train 중에서 FoodCourt 의 값이 null인 경우의 데이터프레임
df_FoodCourt_null = df_FoodCourt.drop('FoodCourt',axis=1)
df_FoodCourt_null
```

```python
# 훈련된 xgboost모델을 사용해서 df_homeplanet_null의 HomePlanet값을 예측
pred = model.predict(df_FoodCourt_null.drop(['PassengerId','Name'],axis=1))
pred.max()
pred = np.round(pred)
```

```python

# 예측된 값을 다시 df_homeplanet으로 삽입(null값 채우기)
df_FoodCourt['FoodCourt'] = pred
df_FoodCourt
```

```python

# null값을 채운 dataframe과 df_train01 합치기 
df_update = pd.concat([df_train01,df_FoodCourt],axis=0)
df_update
```
