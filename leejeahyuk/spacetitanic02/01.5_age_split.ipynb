```python
#과금 그룹화
df = pd.read_csv('./kaggle/input/spaceship-titanic/df_concat.csv')
df.insert(14, 'TotalSpendGroup', df['TotalSpend'])

df.loc[df['TotalSpendGroup']==0,'TotalSpendGroup'] = 0
df.loc[(df['TotalSpendGroup']>0)&(df['TotalSpendGroup']<=1000),'TotalSpendGroup'] = 1000
df.loc[(df['TotalSpendGroup']>1000)&(df['TotalSpendGroup']<=2500),'TotalSpendGroup'] = 2500
df.loc[(df['TotalSpendGroup']>2500)&(df['TotalSpendGroup']<=5000),'TotalSpendGroup'] = 5000
df.loc[(df['TotalSpendGroup']>5000)&(df['TotalSpendGroup']<=10000),'TotalSpendGroup'] = 10000
df.loc[(df['TotalSpendGroup']>10000)&(df['TotalSpendGroup']<=15000),'TotalSpendGroup'] = 15000
df.loc[(df['TotalSpendGroup']>15000),'TotalSpendGroup'] = 15000
```

```python
df.columns
```

```python
# 과금 그래프
import matplotlib.pyplot as plt
import seaborn as sns
spendcount = df.groupby('TotalSpendGroup').count()['PassengerId']
spendcount.shape
spendcount.index.shape
plt.figure(figsize=(15,4))
g = sns.barplot(x=spendcount.index, y=spendcount.values)
g.set_xticklabels(g.get_xticklabels(), rotation=90)
plt.show
```

```python
# 나이 그래프
import matplotlib.pyplot as plt
import seaborn as sns
agecount = df.groupby('Age').count()['PassengerId']
agecount.shape
agecount.index.shape
plt.figure(figsize=(15,4))
g = sns.barplot(x=agecount.index, y=agecount.values)
g.set_xticklabels(g.get_xticklabels(), rotation=90)

plt.title("age distribution")
plt.show
```

```python
import warnings
import seaborn as sns
#Visualizing people of different ages travelled from TotalSpend
plt.figure(figsize=(15,4))
warnings.filterwarnings('ignore')
Age_nomiss=df[~(df.Age.isna())]
Age_nomiss['Age']=Age_nomiss.Age.astype('int')
g=sns.countplot(data=Age_nomiss, x='Age', hue='TotalSpendGroup')
g.set_xticklabels(g.get_xticklabels(), rotation=90)
plt.title("Visualizing people of different ages spend from specetitanic")
plt.show()
```

```python
df = pd.read_csv('./kaggle/input/spaceship-titanic/df_concat.csv')

df.insert(6, 'AgeGroup', df['Age'])

df.loc[df['Age']<=12,'AgeGroup'] = 0
df.loc[(df['Age']>12)&(df['Age']<=17),'AgeGroup'] = 1
df.loc[(df['Age']>17)&(df['Age']<=25),'AgeGroup'] = 2
df.loc[(df['Age'] > 25) & (df['Age'] <= 30),'AgeGroup'] = 3
df.loc[(df['Age'] > 30) & (df['Age'] <= 50),'AgeGroup'] = 4
df.loc[(df['Age'] > 50),'AgeGroup'] = 5

df.to_csv('./kaggle/input/spaceship-titanic/concat_AgeGroup.csv',index=False)
```
